```python
from sklearn.datasets import load_diabetes
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = load_diabetes()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns = df.feature_names
```

```python
dataset.head
```

```python
X = dataset
Y = df.target
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train =scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train,Y_train)
```

```python
mse = cross_val_score(regression, X_train, Y_train, scoring = 'neg_mean_squared_error',cv =10)

np.mean(mse)
```

```python
reg_pred = regression.predict(X_test)
reg_pred
```

```python
import seaborn as sns
sns.displot(reg_pred - Y_test, kind ='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(reg_pred, Y_test)
score
```

# RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge_regression = Ridge()
```

```python
ridge_regression
```

```python
parameters = {'alpha' : [10, 30, 40, 50, 60, 70 ,100, 150, 250, 300]}
ridgecv = GridSearchCV(ridge_regression,parameters,scoring = 'neg_mean_squared_error',cv =10)
ridgecv.fit(X_train, Y_train)
```

```python
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```

```python
ridge_pred = ridgecv.predict(X_test)
ridge_pred
```

```python
sns.displot(ridge_pred - Y_test, kind ='kde')
```

```python
score = r2_score(ridge_pred, Y_test)
score
```

Lasso Regression****

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso_regression = Lasso()
```

```python
lasso_regression
```

```python
parameters = {'alpha' : [10, 30, 40, 50, 60, 70 ,100, 150, 250, 300]}
lassocv = GridSearchCV(lasso_regression,parameters,scoring = 'neg_mean_squared_error',cv =10)
lassocv.fit(X_train, Y_train)
```

```python
print(lassocv.best_params_)
print(lassocv.best_score_)
lasso_pred = lassocv.predict(X_test)
lasso_pred

sns.displot(lasso_pred - Y_test, kind ='kde')

score = r2_score(lasso_pred, Y_test)
score
```
